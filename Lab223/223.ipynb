```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
ly = 579.07*10**(-9)
lg = 546.07*10**(-9)
```

```python
m = np.linspace(1, 20, 20)
```

```python
rsg = np.array([19, 28, 34, 39, 44, 49, 52, 56, 59, 62, 65, 68, 71, 73, 76, 79, 81, 83, 85, 88])*10**(-5) #светлые зеленые
rsy = np.array([18, 28, 34, 40, 45, 49, 53, 57, 60, 63, 66, 69, 72, 75, 78, 80, 83, 85, 87, 89])*10**(-5) #светлые желтые
rtg = np.array([24, 31, 37, 42, 46, 50, 54, 58, 61, 64, 67, 69, 72, 75, 78, 80, 82, 84, 86, 89])*10**(-5) #темные зеленые
rty = np.array([24, 31, 37, 42, 47, 51, 55, 58, 62, 65, 68, 71, 73, 76, 79, 81, 84, 86, 88, 90])*10**(-5) #темные желтые
```

```python
fig, axs = plt.subplots(1, 2, figsize = (15,7))
x = np.linspace(0, 21, 2)
axs[0].set_title("Желтый цвет: светлые кольца Ньютона r^2(m)")
axs[1].set_title("Желтый цвет: темные кольца Ньютона (r')^2(m)")
axs[0].set_xlabel("m, номер полосы")
axs[0].set_ylabel("r^2(m), мм^2")
axs[1].set_xlabel("m, номер полосы")
axs[1].set_ylabel("(r')^2(m), мм^2")
axs[0].scatter(m, (rsy*10**3)**2, s = 150, marker = "|", linewidths = 2, color = "black")
axs[1].scatter(m, (rty*10**3)**2, s = 150, marker = "|", linewidths = 2, color = "black")
k, b = np.polyfit(m, rsy**2, 1)
y = k*x+b
axs[0].plot(x, y*10**6)
ks = k
print(f"ks = {ks}")
k, b = np.polyfit(m, rty**2, 1)
y = k*x+b
kt = k
print(f"kt = {kt}")
k = (ks+kt)/2
print(f"k = {k}")
axs[1].plot(x, y*10**6)
#axs[0].xaxis.set_major_locator(MultipleLocator(1))
#axs[0].yaxis.set_major_locator(MultipleLocator(50))
axs[0].xaxis.set_minor_locator(AutoMinorLocator(5))
axs[0].yaxis.set_minor_locator(AutoMinorLocator(10))
axs[0].grid(which = 'minor', linestyle='--', linewidth = 0.2)
axs[0].grid(which = 'major', linewidth = 0.4)
axs[1].xaxis.set_minor_locator(AutoMinorLocator(5))
axs[1].yaxis.set_minor_locator(AutoMinorLocator(10))
axs[1].grid(which = 'minor', linestyle='--', linewidth = 0.2)
axs[1].grid(which = 'major', linewidth = 0.4)
axs[0].set_xlim(-1, 22)
axs[0].set_ylim(-0.04, 0.9)
axs[1].set_xlim(-1, 22)
axs[1].set_ylim(-0.04, 0.9)
ry = k/ly
print(ry)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (15,7))
x = np.linspace(0, 21, 2)
axs[0].set_title("Зеленый цвет: светлые кольца Ньютона r^2(m)")
axs[1].set_title("Зеленый цвет: темные кольца Ньютона (r')^2(m)")
axs[0].set_xlabel("m, номер полосы")
axs[0].set_ylabel("r^2(m), мм^2")
axs[1].set_xlabel("m, номер полосы")
axs[1].set_ylabel("(r')^2(m), мм^2")
axs[0].scatter(m, (rsg*10**3)**2, s = 150, marker = "|", linewidths = 2, color = "black")
axs[1].scatter(m, (rtg*10**3)**2, s = 150, marker = "|", linewidths = 2, color = "black")
k, b = np.polyfit(m, rsg**2, 1)
y = k*x+b
axs[0].plot(x, y*10**6)
ks = k
print(f"ks = {ks}")
k, b = np.polyfit(m, rtg**2, 1)
y = k*x+b
kt = k
print(f"kt = {kt}")
k = (ks+kt)/2
print(f"k = {k}")
axs[1].plot(x, y*10**6)
axs[0].yaxis.set_major_locator(MultipleLocator(0.2))
axs[1].yaxis.set_major_locator(MultipleLocator(0.2))
axs[0].xaxis.set_minor_locator(AutoMinorLocator(5))
axs[0].yaxis.set_minor_locator(AutoMinorLocator(10))
axs[0].grid(which = 'minor', linestyle='--', linewidth = 0.2)
axs[0].grid(which = 'major', linewidth = 0.4)
axs[1].xaxis.set_minor_locator(AutoMinorLocator(5))
axs[1].yaxis.set_minor_locator(AutoMinorLocator(10))
axs[1].grid(which = 'minor', linestyle='--', linewidth = 0.2)
axs[1].grid(which = 'major', linewidth = 0.4)
axs[0].set_xlim(-1, 22)
axs[0].set_ylim(-0.04, 0.9)
axs[1].set_xlim(-1, 22)
axs[1].set_ylim(-0.04, 0.9)
rg = k/lg
print(rg)
r = (rg+ry)/2
```

```python
srsr = np.array([25, 34, 40, 45, 50, 54, 58, 61, 65, 69])*10**(-5)
srtr = np.array([30, 37, 42, 48, 52, 56, 60, 63, 67, 70])*10**(-5)
srsb = np.array([24, 32, 38, 42, 46, 49, 53, 56, 59, 61])*10**(-5)
srtb = np.array([29, 35, 40, 44, 48, 51, 54, 57, 60, 63])*10**(-5)
```

```python
m = np.linspace(1, 10, 10)
fig, axs = plt.subplots(1, 2, figsize = (15,7))
x = np.linspace(0, 21, 2)
axs[0].set_title("Красный светодиод: светлые кольца Ньютона r^2(m)")
axs[1].set_title("Красный светодиод: темные кольца Ньютона (r')^2(m)")
axs[0].set_xlabel("m, номер полосы")
axs[0].set_ylabel("r^2(m), мм^2")
axs[1].set_xlabel("m, номер полосы")
axs[1].set_ylabel("(r')^2(m), мм^2")
axs[0].scatter(m, (srsr*10**3)**2, s = 150, marker = "|", linewidths = 2, color = "black")
axs[1].scatter(m, (srtr*10**3)**2, s = 150, marker = "|", linewidths = 2, color = "black")
k, b = np.polyfit(m, srsr**2, 1)
y = k*x+b
axs[0].plot(x, y*10**6)
ks = k
print(f"ks = {ks}")
k, b = np.polyfit(m, srtr**2, 1)
y = k*x+b
kt = k
print(f"kt = {kt}")
k = (ks+kt)/2
print(f"k = {k}")
axs[1].plot(x, y*10**6)
axs[0].yaxis.set_major_locator(MultipleLocator(0.1))
axs[1].yaxis.set_major_locator(MultipleLocator(0.1))
axs[0].xaxis.set_major_locator(MultipleLocator(1))
axs[1].xaxis.set_major_locator(MultipleLocator(1))
axs[0].yaxis.set_minor_locator(AutoMinorLocator(10))
axs[0].grid(which = 'minor', linestyle='--', linewidth = 0.2)
axs[0].grid(which = 'major', linewidth = 0.4)
axs[1].yaxis.set_minor_locator(AutoMinorLocator(10))
axs[1].grid(which = 'minor', linestyle='--', linewidth = 0.2)
axs[1].grid(which = 'major', linewidth = 0.4)
axs[0].set_xlim(-1, 11)
axs[0].set_ylim(-0.04, 0.6)
axs[1].set_xlim(-1, 11)
axs[1].set_ylim(-0.04, 0.6)
lr = k/r
print(f"lr = {lr*10**9} мкм")
```
