```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
wl = 546.1 * 10**(-9)
f1 = 0.11
f2 = 0.17
```

```python
n = range(1, 6)
z1 = np.array([3.4, 2.3, 1.8, 1.5, 1.2])*10**(-3)
z2 = np.array([7.3, 4.7, 3.5, 2.6, 2.0])*10**(-3)
z3 = np.array([9.7, 7.9, 4.8, 3.6, 3.2])*10**(-3)
ksi1 = 2*(z1*n*wl)**0.5
ksi2 = 2*(z2*n*wl)**0.5
ksi3 = 2*(z3*n*wl)**0.5
print(ksi1, ksi2, ksi3)
```

```python
ls = np.array(["b = 0.1 мм", "b = 0.15 мм", "b = 0.2 мм"])
ksi = [ksi1, ksi2, ksi3]
m = ["+", "D", "2"]
sz = [45, 15, 45]
fig, axs = plt.subplots(1,1, figsize = (10,10))
for i in range(0, 3):
    axs.scatter(n, ksi[i]*10**3, marker = m[i], s = sz[i], label = ls[i], color = "black")
axs.set_ylim(0, 0.25)
axs.xaxis.set_major_locator(MultipleLocator(1))
axs.xaxis.set_minor_locator(AutoMinorLocator(10))
axs.yaxis.set_minor_locator(AutoMinorLocator(10))
plt.grid(which = 'minor', linestyle='--', linewidth = 0.4)
plt.grid(which = 'major', linewidth = 0.6)
axs.set_title("Ширина щели от количества открытых зон Френеля")
axs.set_xlabel("Открытые зоны")
axs.set_ylabel("Ширина щели, мм")
axs.legend()
plt.show()
```

```python
n = range(0, 4)
x = np.linspace(0, 3, 10)
dx1 = np.array([0, 0.1, 0.2, 0.3])
dx2 = np.array([0, 0.2, 0.3, 0.4])
dx3 = np.array([0, 0.3, 0.6, 0.9])
dx = [dx1, dx2, dx3]
ls = np.array(["b = 1 мм", "b = 0.65 мм", "b = 0.3 мм"])
m = ["+", "D", "2"]
sz = [45, 15, 45]
fig, axs = plt.subplots(1,1, figsize = (10,10))
kss = []
for i in range(0, 3):
    axs.scatter(n, dx[i], marker = m[i], s = sz[i], label = ls[i], color = "black")
    k,b = np.polyfit(n, dx[i], 1)
    kss.append(k)
    y = k*x + b
    axs.plot(x, y)
ks = np.array(kss)*10**(-3)
print(kss)
b = wl*f2/ks
print(b)
axs.xaxis.set_major_locator(MultipleLocator(1))
axs.xaxis.set_minor_locator(AutoMinorLocator(10))
axs.yaxis.set_minor_locator(AutoMinorLocator(10))
plt.grid(which = 'minor', linestyle='--', linewidth = 0.4)
plt.grid(which = 'major', linewidth = 0.6)
axs.set_title("Расположение минимумов дифракции Фраунгофера")
axs.set_xlabel("Номер минимума")
axs.set_ylabel("Расстояние до минимума, мм")
axs.legend()
plt.show()
```
