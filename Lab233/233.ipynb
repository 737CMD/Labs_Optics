```python
lb = np.array(["", "", " ", ""])
xp = [np.array([19.45, 20, 21, 22, 23, 24, 25, 26, 27, 27.9])-19,
     np.linspace(0, 10, 11),
     np.linspace(0, 11, 12),
     np.linspace(0, 11, 12)]
yp = [np.array([0, 2, 10, 16, 19, 19, 17.5, 12.5, 5, 0])/2,
         np.array([16, 12.5, 19, 28, 35.5, 38.5, 36, 29, 19, 12.5, 14])/2, 
         np.array([24, 20, 22, 28, 34, 36, 36, 32, 26, 21, 20, 25])/2,
         np.array([26, 25, 25, 27, 29.5, 32, 32, 30.5, 28, 25, 24.5, 25.5])/2]
m = ["+", "x", "+", "x"]
sz = [45, 35, 45, 35]
fig, axs = plt.subplots(2,1, figsize = (15,20), gridspec_kw={'height_ratios': [2, 1]})
for i in range(0, 2):
    axs[0].scatter(xp[i], yp[i], marker = m[i], s = sz[i], label = lb[i], color = "black")
for i in range(2, 4):
    axs[1].scatter(xp[i], yp[i], marker = m[i], s = sz[i], label = lb[i], color = "black")
axs[0].xaxis.set_major_locator(MultipleLocator(1))
axs[0].yaxis.set_major_locator(MultipleLocator(2.5))
axs[0].xaxis.set_minor_locator(AutoMinorLocator(10))
axs[0].yaxis.set_minor_locator(AutoMinorLocator(10))
axs[1].xaxis.set_major_locator(MultipleLocator(1))
axs[1].yaxis.set_major_locator(MultipleLocator(2.5))
axs[1].xaxis.set_minor_locator(AutoMinorLocator(10))
axs[1].yaxis.set_minor_locator(AutoMinorLocator(10))
axs[0].grid(which = 'minor', linestyle='--', linewidth = 0.4)
axs[0].grid(which = 'major', linewidth = 0.6)
axs[1].grid(which = 'minor', linestyle='--', linewidth = 0.4)
axs[1].grid(which = 'major', linewidth = 0.6)
axs[0].set_title("      ")
axs[0].set_xlabel("  ")
axs[1].set_xlabel("  ")
axs[0].set_ylabel(" ")
axs[1].set_ylabel(" ")
axs[0].set_xlim(-0.5, 11.5)
axs[1].set_xlim(-0.5, 11.5)
axs[0].set_ylim(-1, 20)
axs[1].set_ylim(7.5, 20)
axs[0].legend()
axs[1].legend()
plt.show()
```
