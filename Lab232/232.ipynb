```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
from scipy import interpolate
```

```python
#поляризатор + анализатор
theta = np.linspace(0, np.pi, 19)
teta = np.linspace(0, np.pi, 200)
r = np.array([2, 67, 232, 477, 784, 1109, 1395, 1632, 1776, 1824, 1766, 1598, 1349, 1030, 733, 426, 190, 45, 2])
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(projection='polar')
ax.scatter(theta, r, marker='+', color = "black", s = 45, label = "Экспериментальные точки")
ax.plot(teta, 1824*np.cos(teta-np.pi/2)**2, label = "Imax * cos^2(φ - 90°)")
ax.set_title("Проверка закона Малюса")
ax.set_ylabel("Освещенность, Люксы ")
ax.set_xlim(0, np.pi)
ax.set_ylim(0,2000)
plt.legend()
plt.plot()
```

```python
#фазовая пластинка установлена под 45 к главным осям
r = np.array([12, 54, 172, 365, 608, 862, 1103, 1300, 1436, 1495, 1456, 1342, 1158, 905, 640, 395, 192, 67, 12])
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(projection='polar')
ax.scatter(theta, r, marker='+', color = "black", s = 45, label = "Экспериментальные точки")
ax.plot(teta, 1495*np.cos(teta-np.pi/2)**2, label = "Imax * cos^2(φ - 90°)")
ax.set_title("Фазовая пластинка установлена под 45° к главным осям между скрещенным поляризатором и анализатором ")
ax.set_ylabel("Освещенность, Люксы ")
ax.set_xlim(0, np.pi)
ax.set_ylim(0,1600)
plt.legend()
plt.plot()
```

```python
#Экперименты с черным зеркалом
r = np.array([10, 19, 38, 70, 110 ,150, 190, 220, 239, 244, 237, 217, 185, 150, 109, 70, 39, 18, 10])
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(projection='polar')
ax.scatter(theta, r, marker='+', color = "black", s = 45, label = "Экспериментальные точки")
ax.plot(teta, 244*np.cos(teta-np.pi/2)**2, label = "Imax * cos^2(φ - 90°)")
ax.set_title("Интенсивность отраженного при угле падения 50° света от угла поворота поляризатора")
ax.set_ylabel("Освещенность, Люксы ")
ax.set_xlim(0, np.pi)
ax.set_ylim(0,250)
plt.legend()
plt.plot()
```

```python
#Экперименты с черным зеркалом
theta = np.linspace(15, 80, 14)
thetas = np.linspace(15, 80, 100)
v = np.array([61, 57,53, 47, 40, 33, 23, 10, 3, 5 ,29, 102, 235, 530])
g = np.array([74, 85, 98, 111, 126, 155, 189, 246, 323, 399, 530, 678, 899, 1300])
fig, ax = plt.subplots(1, 1, figsize = (10,10))
ax.scatter(np.insert(theta, 9, 57), np.insert(v, 9, 1), marker='+', color = "black", s = 45, label = "Вертикальная поляризация")
ax.scatter(theta, g, marker='x', color = "black", s = 45, label = "Горизонтальная поляризация")
vy = interpolate.Akima1DInterpolator(np.insert(theta, 9, 57), np.insert(v, 9, 1))(thetas)
ax.plot(thetas, vy, label="Интерполяция вертикальной поляризации")
gy = interpolate.Akima1DInterpolator(theta, g)(thetas)
ax.plot(thetas, gy, label="Интерполяция горизонтальной поляризации")
ax.set_title("Зависимость отраженного света интенсивности от угла падения")
ax.set_ylabel("Освещенность, Люксы ")
ax.set_xlim(0, 90)
ax.set_ylim(-5, 1350)
ax.xaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_major_locator(MultipleLocator(100))
ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(10))
plt.grid(which = 'minor', linestyle='--', linewidth = 0.2)
plt.grid(which = 'major', linewidth = 0.4)
plt.legend()
plt.plot()
```
