```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
```

```python
interftabl = [[-330, -330, -325, -325, -335, -335, -330, -330], 
              [-250, -270, -260, -270, -265, -265, -260, -260], 
              [-190, -200, -190, -210, -190, -200, -193, -192],
              [-125, -135, -130, -130, -135, -135, -125, -136],
              [-75, -65, -57, -65, -55, -66, -55, -75],
              [0],
              [60, 80, 65, 75, 60, 63, 61, 70],
              [120, 135, 125, 135, 125, 130, 127, 133],
              [195, 195, 195, 200, 180, 185, 195, 195],
              [270, 270, 260, 270, 265, 265, 270, 265],
              [330, 330, 330, 330, 330, 330, 330, 330]]
interf = np.array([np.average(i) for i in interftabl])
sigm = np.array([np.std(i) for i in interftabl])
sigm
```

```python
xpoints = np.linspace(-5, 5, 11)
x = np.linspace(-5, 5, 110)
fig, axs = plt.subplots(1,1, figsize = (10,10))
axs.set_title("     N(deltaP)")
axs.set_xlabel("N, ")
axs.set_ylabel(" , . . .")
axs.scatter(xpoints, interf, s = [15, 35, 35, 22 ,40, 10, 40, 30, 35, 15, 10], marker = "|", linewidths = 2, color = "black")
k, b = np.polyfit(xpoints, interf, 1)
y = k*x+b
axs.plot(x, y)
axs.xaxis.set_major_locator(MultipleLocator(1))
axs.yaxis.set_major_locator(MultipleLocator(50))
#axs.xaxis.set_minor_locator(AutoMinorLocator(5))
axs.yaxis.set_minor_locator(AutoMinorLocator(5))
plt.grid(which = 'minor', linestyle='--', linewidth = 0.2)
plt.grid(which = 'major', linewidth = 0.4)
```
