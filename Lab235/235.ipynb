```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
from scipy import interpolate
```

```python
j = np.array([44*20, 38*20, 26*20, 17*20, 17*10, 8*10, 11*2, 6])*10**(-6)
jp = j/j[0]
r = np.linspace(0.2, 1.6, 8)*10**(-3)
fig, axs = plt.subplots(1,1, figsize = (10, 9))
axs.scatter(r*1000, jp, marker = "+", s = 45, color = "black")
axs.xaxis.set_major_locator(MultipleLocator(0.1))
axs.xaxis.set_minor_locator(AutoMinorLocator(10))
axs.yaxis.set_major_locator(MultipleLocator(0.1))
axs.yaxis.set_minor_locator(AutoMinorLocator(10))
plt.grid(which = 'minor', linestyle='--', linewidth = 0.4)
plt.grid(which = 'major', linewidth = 0.8)
axs.set_title("   ")
axs.set_xlabel(", ")
axs.set_ylabel(" ")
x = np.linspace(0.2, 1.6, 100)
cubic = interpolate.Akima1DInterpolator(r*1000, jp)
y = cubic(x)
axs.plot(x, y)
plt.show()
```
